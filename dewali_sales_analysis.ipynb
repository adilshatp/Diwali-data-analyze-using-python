```python
# here we have 12 null values in the amount column.
# removing the null values
```

# Explonatory Data Analysis

### Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gg=df.groupby(['Gender'], as_index=False)['Amount'].sum()
gg
```

```python
sns.barplot(x='Gender',y='Amount', data= gg)
```

### AGE

```python
aa=sns.countplot(data=df, x='Age Group', hue = 'Gender')
for bars in aa.containers:
    aa.bar_label(bars)
```

```python
# Total Amount VS Age Group
sale_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum()
sns.barplot(x='Age Group', y='Amount', data = sale_age)
```

From above graphs we can see that most of the buyers of age group between 26-33 yrs female

### State

```python
df.columns
```

```python
# total number of order from top states
```

```python
tt=df.groupby(['State'], as_index=False)['Orders'].sum().head(10)
tt
```

```python
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=tt, x='State', y='Orders')
```

From above graph we can see that most of the orders are from uttar pradesh, maharashtra, and karnataka

### Martial Status

```python
ax = sns.countplot(data=df, x='Marital_Status')

sns.set(rc={'figure.figsize':(8,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
mm=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
mm
```

```python
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=mm, x='Marital_Status',y='Amount', hue='Gender')
```

From the abouve graph, we can see that most of the buyers are married women

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
oc=sns.countplot(data=df, x='Occupation')

for bars in oc.containers:
    oc.bar_label(bars)
```

From the above graph , the most of the buyers are working in IT sectors

```python
df.columns
```

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
pc=sns.countplot(data=df, x='Product_Category')

for bars in pc.containers:
    pc.bar_label(bars)
```

```python
pp=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
pp
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=pp, x='Product_Category',y='Amount')
```

```python
# Top 3
```

```python
oo=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(3)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=oo, x='Product_Category',y='Amount')
```

```python
df.columns
```

```python
# Top selling product
```

```python
pi=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=pi, x='Product_ID', y='Orders')
```
